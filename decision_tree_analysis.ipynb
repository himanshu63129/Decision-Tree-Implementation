# Decision Tree Classification: Iris Dataset

This notebook demonstrates building, visualizing, and analyzing a Decision Tree model using the classic Iris dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Prettify plots
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100
sns.set_style("whitegrid")
```

## 1. Load and Explore the Data

We use the Iris dataset, which contains measurements of 150 iris flowers from three different species.

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

print(f"Dataset Shape: {df.shape}")
df.head()
```

## 2. Prepare the Data

We'll split the data into training and testing sets to evaluate the model's performance on unseen data.

```python
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training instances: {X_train.shape[0]}")
print(f"Testing instances: {X_test.shape[0]}")
```

## 3. Train the Decision Tree Model

We use a `DecisionTreeClassifier` with default parameters, except for `random_state` for reproducibility.

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

## 4. Visualize the Decision Tree

Visualizing the tree helps us understand the decision-making process of the model.

```python
plt.figure(figsize=(20,10))
plot_tree(clf, 
          filled=True, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names, 
          rounded=True, 
          fontsize=12)
plt.title("Decision Tree Visualization - Iris Dataset")
plt.show()
```

## 5. Model Evaluation and Analysis

Now we evaluate how well the model predicts the iris species on the test set.

```python
y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy:.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# Confusion Matrix Heatmap
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## 6. Feature Importance

Which features were most important for the decisions?

```python
importances = clf.feature_importances_
feature_imp_df = pd.DataFrame({'Feature': iris.feature_names, 'Importance': importances})
feature_imp_df = feature_imp_df.sort_values(by='Importance', ascending=False)

sns.barplot(x='Importance', y='Feature', data=feature_imp_df, palette='viridis')
plt.title('Feature Importance in Decision Tree')
plt.show()
```
